```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import AnovaRM
```

```python
# Load data into a pandas DataFrame
df = pd.read_csv(r'C:\Users\komal ramesh sapkal\Dropbox\Project\anovarm_data.csv')
print(df.to_latex())
```

```python
# Define the repeated measures ANOVA model
model = ols('NetGeneration ~ C(Power_Station)*C(Month)', data=df).fit()

# Compute the repeated measures ANOVA table
aov_table = AnovaRM(df, 'NetGeneration', 'Power_Station', within=['Month'],aggregate_func='mean').fit()

# Print the ANOVA table
print(aov_table.summary())
```
