```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'C:\Users\FLUTE\Dropbox\Project\TPPS_Perf.xlsx')
```

```python
df1=df[[ 'MUG', 'PLFC', 'AFC', 'L', 'PLFM', 'AVFM', 'CC', 'AC', 'NG']]
df1.head()
```

# Multicollinearity Test

```python
print(df1.corr())
```

# Linearity and Normality Test

```python
import seaborn as sns
sns.set(style='ticks',color_codes=True,font_scale=2)
g=sns.pairplot(df1,height=3,diag_kind='hist',kind='reg')
g.fig.suptitle('Scatter Plot',y=1.08)
```

# Homoscedasticity

```python
from statsmodels.formula.api import ols
reg=ols('NG~MUG + PLFC + AFC + L + PLFM + AVFM + CC + AC',data=df1).fit()
print(reg.summary())
```

```python
fig,ax=plt.subplots(figsize=(8,3.5))
pred_val=reg.fittedvalues.copy()
true_val=df1['NG'].values.copy()
resid=true_val-pred_val
res=sns.residplot(resid,pred_val)
plt.title('Homoscedasticity')
```

# Normality of Errors

```python
import scipy.stats as stats
fig,ax=plt.subplots(figsize=(10,6))
stats.probplot(resid,dist='norm',plot=plt)
plt.show
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X=df1[['PLFC','AFC']]
y=df1[['L']]
```

```python
# Create a model and fit it to the data
model = sm.OLS(y, X)
results = model.fit()
```

```python
# Print the summary of the regression results
print(results.summary())
```

```python
# Get the estimated coefficients
coefficients = results.params
print("Coefficients:")
print(coefficients)
```

```python
# Make predictions using the model
new_X = sm.add_constant(X)  # Assuming you have new data in `new_X`
```

```python
predictions = results.predict(new_X)
print("Predictions:")
print(predictions)
```
